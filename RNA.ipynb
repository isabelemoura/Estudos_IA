```python
import readline

class StandardInputLine:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class InputLine(StandardInputLine):
    def __init__(self, x, y, w, bias=None, wBias=None):
        super().__init__(x, y)
        self.w = w
        self.bias = bias
        self.wBias = wBias

class Line:
    def __init__(self, inputLine, fnet, output, diff, delt, deltBias=None):
        self.inputLine = inputLine
        self.fnet = fnet
        self.output = output
        self.diff = diff
        self.delt = delt
        self.deltBias = deltBias

def calc_fnet(inputLine):
    result = sum(x * w for x, w in zip(inputLine.x, inputLine.w))
    if inputLine.bias is not None and inputLine.wBias is not None:
        result += inputLine.bias * inputLine.wBias
    return result

def calc_output(fnet):
    return 1 if fnet > 0 else 0

def calc_diff(output, y):
    return y - output

def calc_delt(rate, diff, value):
    return rate * diff * value

def calc_line(StandardInputLine, w, rate, bias=None, wBias=None):
    inputLine = InputLine(StandardInputLine.x, StandardInputLine.y, w, bias, wBias)
    fnet = calc_fnet(inputLine)
    output = calc_output(fnet)
    diff = calc_diff(output, inputLine.y)
    delt = [calc_delt(rate, diff, x) for x in inputLine.x]
    deltBias = calc_delt(rate, diff, inputLine.bias) if inputLine.bias is not None else None
    return Line(inputLine, fnet, output, diff, delt, deltBias)

def is_finished_cycle(cycle):
    return all(line.inputLine.y == line.output for line in cycle)

def calc_perceptron(standardInputLines, initialW, rate, bias=None, wBias=None):
    cycles = []
    currentW = initialW[:]
    currentWBias = wBias

    while True:
        lines = []
        for stdLine in standardInputLines:
            line = calc_line(stdLine, currentW, rate, bias, currentWBias)
            currentW = [w + dw for w, dw in zip(line.inputLine.w, line.delt)]
            if line.inputLine.wBias is not None and line.deltBias is not None:
                currentWBias = line.inputLine.wBias + line.deltBias
            lines.append(line)
        cycles.append(lines)
        if is_finished_cycle(cycles[-1]):
            break
    return cycles

def get_final_weights(cycles):
    finishedCycle = cycles[-1]
    line = finishedCycle[0]
    return line.inputLine

def input_perceptron(w=None, wB=None):
    while True:
        stdLines = []
        for i in range(3):
            x = list(map(float, input(f"Insira x1, x2, x3 (separados por virgula) para o par {i + 1}: ").split(',')))
            y = float(input(f"Insira y (valor esperado) para o par {i + 1}: "))
            stdLines.append(StandardInputLine(x, y))
        initialW = w or list(map(float, input("Insira os pesos iniciais w1, w2, w3 (separados por virgula ','): ").split(',')))
        bias = float(input("Caso possua Bias informe o valor do mesmo (caso nao possua coloque o valor 0): "))
        if wB is None:
            wBias = float(input("Caso possua Bias informe o valor do seu peso (caso nao possua coloque o valor 0): ").replace(',', '.'))
        else:
            wBias = wB
        rate = float(input("Informe o valor da taxa de aprendizagem: "))

        cycles = calc_perceptron(stdLines, initialW, rate, bias, wBias)
        display(cycles)

        if input("\nDeseja inserir um novo treinamento utilizando os pesos do último treinamento? (sim/nao): ").strip().lower() != 'sim':
            break
        else:
            final_weights = get_final_weights(cycles)
            w = final_weights.w
            wB = final_weights.wBias

def display(cycles):
    line = get_final_weights(cycles)
    print(f"\n\nQuantidade de iterações necessárias: {len(cycles)}")
    print(f"Pesos aprendidos: {line.w + [line.wBias]}")

def main():
    input_perceptron()

if __name__ == "__main__":
    main()
4
```
